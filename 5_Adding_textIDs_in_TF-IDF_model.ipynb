## Loading the required libraries

```python
import json
```

## Reading tf-idf modeled corpus line by line

```python
tfidf_data = []
for line in open('drive/MyDrive/Colab Notebooks/tfidf_texts.json', 'r'):
    tfidf_data.append(json.loads(line))
```

```python
print(tfidf_data[0])
```

## Adding IDs of texts in tf-idf model. Saving the result to a json file

```python
with open('drive/MyDrive/Colab Notebooks/tfidf_data_withID.json', 'w') as f:
    i = 0
    for text in tfidf_data:
        temp_list = [['text%s' %i, word[0], word[1]] for word in text]
        i += 1
        f.write(json.dumps(temp_list) + '\n')
```
