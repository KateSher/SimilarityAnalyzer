## Import Library similarity_analyzer

```python
from similarity_analyzer import RelevantText
```

## Loading a model and a text document
The [model](https://drive.google.com/file/d/1EfPDXJPvU4xli1mnJznwgusOuAoTUmpW/view?usp=sharing) and the [text file](https://drive.google.com/file/d/1JEADRYcLNeuIO3-k4L7Uk9slzl22Tnmx/view?usp=sharing) are on google drive.

```python
# Path to model and text file
model_path = 'full_result_dict.pickle'
text_path = 'wiki(num_title).txt'
```

## Creating an instance of a class RelevantText

```python
rt = RelevantText(model_path, text_path)
```

## Getting a list of a given number of IDs of relevant articles for a keyword in accordance with TF-IDF

```python
list_num_texts = rt.text_relevance_by_word('cat', 4)
```

```python
if list_num_texts:
    print('Get a list of relevant articles:', list_num_texts)
else:
    print('Word not found')
```

## Getting relevant articles according to the list of IDs

```python
list_texts = rt.find_text(list_num_texts)
```

```python
for i in list_texts:
    print('\n Article number:', i[0])
    print('\n Title:', i[1])
    print('\n Content:\n\n', i[2])
    print('\n', '*' * 120)
```
