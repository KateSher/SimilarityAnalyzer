## Compiling 4 batches of the dictionary into one whole. 
## Sort keywords alphabetically.

```python
# dictionary part numbers
num_doc = [1, 2, 3, 4]
```

```python
full_result_dict = []
for i in num_doc:  
    with open('drive/MyDrive/Colab Notebooks/result_dict({}).json'.format(i), 'r') as f: 
        dct = json.load(f)
        dct = list(dct.items())
        full_result_dict += dct
# sorting a dictionary by keys in alphabetical order
full_result_dict.sort(key=lambda x: x[0])
full_result_dict = dict(full_result_dict)
```

```python
len(full_result_dict)
```

## Saving the finished dictionary (complete TF-IDF model)

Save to a json file

```python
with open('drive/MyDrive/Colab Notebooks/full_result_dict.json', 'w') as f:
    json.dump(full_result_dict, f, indent=4)
```

Save to a pickle file

```python
with open('drive/MyDrive/Colab Notebooks/full_result_dict.pickle', 'wb') as f:
    pickle.dump(full_result_dict, f, protocol=pickle.HIGHEST_PROTOCOL)
```
