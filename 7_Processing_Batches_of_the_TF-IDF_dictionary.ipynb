## Sorting and cutting up to 30 values of each dictionary (from 4 folders).
## Saving 4 batches of the dictionary.

Saving dictionaries for each keyword with its tf-idf weight and text ID, sorted by weight descending, and cuted to 30 values (most significant).

```python
# dictionary part numbers
num_doc = [1, 2, 3, 4]
```

```python
# Sorting and cutting up to 30 values of each dictionary (from 4 folders)
for i in num_doc:
    file_list = glob.glob('drive/MyDrive/Colab Notebooks/all_temporary/temporary{}/*.json'.format(i))
    result_dict = {}
    for file_name in file_list:
        f = open(file_name)
        result = f.read()
        f.close()
        result = json.loads(result)
        key =  list(result.keys())[0]
        values = list(result.values())[0]
        values.sort(key=lambda x: x[1], reverse=True)
        values = values[:30]
        result_dict[key] = values

    # Saving 4 batches of the dictionary
    with open('drive/MyDrive/Colab Notebooks/result_dict({}).json'.format(i), 'w') as f:
        json.dump(result_dict, f, indent=4)
```
